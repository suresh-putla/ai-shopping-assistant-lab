## Filter by year 2022

```python
def filter_data_by_date(data: dict) -> bool:
    filter= False
    if int(data["details"]["Date First Available"][-4:]) < 2022:
        filter = True
        return filter
```

```python
with gzip.open("../../data/meta_Electronics.jsonl.gz", "rt") as fp:
    with open("../../data/meta_Electronics.jsonl", "a", encoding= "utf-8") as fp_out:
        with open("../../data/meta_Electronics.jsonl", "a", encoding= "utf-8") as fp_out_no_date:
            i =0
            for line in fp:
                data = json.loads(line.strip())
                try:
                    filter = filter_data_by_date(data)
                    if not filter:
                        json.dump(data, fp_out)
                        fp_out.write('\n')
                        fp_out.flush()
                except:
                    json.dump(data, fp_out_no_date)
                    fp_out_no_date.write('\n')
                    fp_out_no_date.flush()
                i += 1
                if i% 10000 == 0:
                    print(f"processed {i} lines")
```

```python
def filter_category(data:dict) -> dict:
    if data['main_category'] == None:
        return False
    return True
```

```python
with open("../../data/meta_Electronics.jsonl", "rt") as fp:
    with open("../../data/meta_Electronics_with_category.jsonl", "a", encoding= "utf-8") as fp_cat:
        with open("../../data/meta_Electronics_with_no_category.jsonl", "a", encoding= "utf-8") as fp_out_no_cat:
            i =0
            for line in fp:
                data = json.loads(line.strip())
                filter = filter_category(data)
                if filter:
                    json.dump(data, fp_cat)
                    fp_cat.write('\n')
                    fp_cat.flush()
                else:
                    json.dump(data, fp_out_no_cat)
                    fp_out_no_cat.write('\n')
                    fp_out_no_cat.flush()
                i += 1
                if i% 10000 == 0:
                    print(f"processed {i} lines")
```
