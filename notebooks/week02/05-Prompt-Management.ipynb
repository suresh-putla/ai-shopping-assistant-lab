```python
def build_prompt_jinja(context, query):
    jinja_prompt = """
    You are a shopping assistant that can answer questions about the products in stock.

    You will be given a question and a list of context.

    Instructions:
    - answer the question based on the provided context only.
    - never use word context and refer to it as the available products.
    - do not use markdown formatting.
    - if you do not find any products, say 'there is no product in the stock'.

    Context:
    {{ context }}

    Question:
    {{ query }}
    """

    jinja_template = Template(jinja_prompt)
    rendered_template = jinja_template.render(context="dd", query = "x")
    return rendered_template
```

```python
prompt = build_prompt_jinja("xx", "rr")
print(prompt)
```

```python
def prompt_template_config(yaml_path, prompt_key):
    with open(yaml_path, "r") as file:
        config = yaml.safe_load(file)
    
    template_content = config["prompts"][prompt_key]
    template = Template(template_content)
    return template
```

```python
template = prompt_template_config("prompts/retrieval_generation.yml","retrieval_generation")
template
```

```python
rendered_prompt = template.render(context="testcontxt",query="testquery")
print(rendered_prompt)
```

```python
def build_prompt_jinja2(context, query):
    template = prompt_template_config("prompts/retrieval_generation.yml","retrieval_generation")
    rendered_prompt = template.render(context="testcontxt",query="testquery")
    return rendered_prompt
```

```python
prompt = build_prompt_jinja2("textContext", "textQuery")
print(prompt )
```

### Prompt Registries

```python
ls_client = Client()
```

```python
ls_ret_gen = ls_client.pull_prompt("retrieval-generation")
ls_ret_gen
```

```python
ret_gen_prompt = ls_ret_gen.messages[0].prompt.template
ret_gen_prompt
```

```python
def prompt_template_registry(prompt_name):
    ls_ret_gen = ls_client.pull_prompt(prompt_name).messages[0].prompt.template
    return Template(ls_ret_gen)
```

```python
print(prompt_template_registry("retrieval-generation"))
```
