```python
from langsmith import Client
```

### Attach Feedback to a specific trace in LangSmith

```python
client = Client()
```

```python

RUN_ID = "019f762f-f5d8-70a3-9b71-fa20afaa7580"
```

```python

client.create_feedback(
    run_id=RUN_ID,
    key="feedback",
    score=0,
    feedback_source_type="api")
```

```python

client.create_feedback(
    run_id=RUN_ID,
    key="comment",
    value="this is free text feedback",
    feedback_source_type="api")
```

```python

client.create_feedback(
    run_id=RUN_ID,
    key="comment_with_score",
    value="this is free text feedback",
    score=1,
    feedback_source_type="api")
```
